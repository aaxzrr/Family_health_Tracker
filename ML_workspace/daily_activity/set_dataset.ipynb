```python
import pandas as pd
from datetime import datetime
import os
```

```python
def move_to_json(name, path, output, activities_to_keep):
    columns=['user','activity','time','x','y','z']
    
    parts = name.split('_')
    user = parts[1]
    o_file = f'{parts[-2]}_{parts[-1]}'
    
    user_folder = os.path.join(output, user)
    os.makedirs(user_folder, exist_ok=True) 
    output_file = os.path.join(user_folder, f"{o_file}.json")
    
    df = pd.read_csv(path, sep=",", header=None)
    temp=pd.DataFrame(data=df.values, columns=columns)
    temp = temp[temp['activity'].isin(activities_to_keep)]
    temp['time'] = temp['time'].apply(lambda t: datetime.utcfromtimestamp(t / 1_000_000).replace(year=2024).strftime("%Y-%m-%d %H:%M:%S"))   
    temp['z'] = temp['z'].astype(str).str.replace(';', '', regex=False).astype(float)
    temp = temp.drop(['user', 'activity','time'], axis=1)
    # print(temp['z'])
    temp.to_json(output_file, orient='records',lines=False, indent=4)
```

```python
raw_path = r'D:\bangkit\capstone\Family_health_Tracker\ML_workspace\daily_activity\raw'
keep = ['A', 'B', 'C', 'D', 'E', 'M', 'P', 'O', 'Q']
for device in os.listdir(raw_path):
    device_path = os.path.join(raw_path, device)
    if device == '.DS_Store':
        continue
    for move in os.listdir(device_path):
        move_path = os.path.join(device_path, move)
        if move == '.DS_Store':
            continue    
        for file in os.listdir(move_path):
            file_path = os.path.join(move_path, file)
            if file == '.DS_Store':
                continue
            move_to_json(file.split('.')[0],file_path,r'D:\bangkit\capstone\Family_health_Tracker\ML_workspace\daily_activity\database_json',keep)
```
